```python
#lets check the columns name to see if we want to correct that...make sure no spaces in column names
df.columns.to_list()
# we can see campaign_id has whitespaces and all names are capitalize
```

```python
#replace the space in column name
df.columns = df.columns.str.strip().str.replace(" ","_")
#we can  use str.strip() to replace the whitespace before and after name and replace() to change whitespace in between names
```

```python
df.columns.to_list()
```

```python
#make columns names in lower space
df = df.rename(columns = str.lower)
#aalternate use df.columns = df.columns.str.lower()
```

```python
df.columns
```

```python
#check data where it has $ or . or , in spend column as it makes it a string rather than number
df.loc[df.spend.astype(str).str.contains(r'\$.')].head(3)
```

```python
df['spend'] = df.spend.astype(str).str.replace(r'[^\d.-]','',regex=True).head(3)
df.loc[df.spend.astype(str).str.contains(r'\d.')].head(3)
```

```python
df.spend =pd.to_numeric(df.spend)
df.dtypes
```

```python
#10 uniques values for now,it ignores NA values, use dropna=False to count NA
print(df.channel.nunique(dropna=False))
print(df.channel.unique()) #values like TikTok and Tik_Tok are same also Facebook and Facebok..lets fix this
```

```python
#fix channel values
channel_replace = {
    'Tik_Tok':'TikTok',
    'Facebok' :'Facebook',
    'Insta_gram':'Instagram',
    'E-mail':'Email',
    'Gogle':'Google Ads',
    'N/A':np.nan
}
df['channel'] = df.channel.replace(channel_replace) 
print(df.channel.unique())
```

```python
#fix active values and make it boolean
print(df.active.unique())
active_fix = {'Y':True ,
              '0':False ,
              0:False,
              'No':False ,
              'Yes':True ,
              'N':False,
              '1': True,
              1: True
}
df['active'] = df.active.replace(active_fix).fillna(False).astype(bool)
print(df.active.unique())
```

```python
#fix date columns and make it datetime
print(df.start_date.head(5))
print(df.end_date.head(5))
#print(df.dtypes['start_date'])

df['start_date'] = pd.to_datetime(df.start_date,errors= 'coerce')
df['end_date'] = pd.to_datetime(df.end_date,dayfirst=True,errors= 'coerce')
print(df.start_date.head(5))
print(df.end_date.head(5))
```

```python
#delete duplicate columns
df = df.loc[:,~df.columns.duplicated()].copy()
df.head()
```

```python
#checking data integrity in clicks vs impressions as clicks should be smaller than impressions
df[df.clicks > df.impressions]
```

```python
#checking data integrity in dates as start_date <= end_date
wrongddate = df.start_date > df.end_date
df.loc[wrongddate,'end_date'] = df['start_date'] +pd.Timedelta(days=30)
df.loc[wrongddate]
```
